Difference approximation using second order, center difference approximation of the below ODE

\begin{align}
    u''(x) = f(x) \hphantom{rgd}\text{for } 0<x<1
\end{align}
with some given boundary conditions
\begin{align}
    u(0) = \alpha, \hphantom{hello} u(1) = \beta
\end{align}

The difference equation we get after using the approximation $\displaystyle D^2u(x) \equiv \frac{u(x+h) - 2u(x) + u(x-h)}{h^2} $

\begin{align}
    \frac{1}{h^2}\left(U_{j-1} - 2U_j + U_{j+1}\right) = f(x_j) \hphantom{rgd} \text{for } j = 1,2,\ldots,m.
\end{align}
\begin{align}
    U_0 = \alpha, U_{m+1} = \beta
\end{align}

```python
def f(x):
    return(-np.sin(x))

def a_func(x):
    return x/x*1.0

def b_func(x):
    return x*0

def c_func(x):
    return x*0

def tridiagonal(dia, up, bel):
    return(np.diag(dia, 0) + np.diag(up, 1) + np.diag(bel, -1))
```

```python
# Boundary points

# Find a bug and then check the convergence rate
# Found the bug, the bug was somewhere the in the part where rhs is defined

# a = 0.0
# b = 2*np.pi

# # Mesh size
# m = 100
# h = (b-a)/(m+1)

# Grid points
# grid_pts = [a]
# for i in range(1, m + 2): # Since we want to use the previous grid point to get the next one, so we start 
#                             the loop 
#                           # from 1 instead of 0
#     grid_pts.append(grid_pts[i - 1] + h)
    
# print(grid_pts)

def second_ord(a, b, alpha, beta, f, m, tridiagonal):
    h = (b - a)/(m+1)

    grid_pts = np.linspace(a,b, m + 2)

    
    Coeff = tridiagonal(np.full(m, -2), np.full(m - 1, 1), np.full(m - 1, 1))

    F = h*h*f(grid_pts[1:m+1])
    F[0] -= alpha
    F[-1] -= beta

    Num_sol = np.linalg.solve(Coeff, F)
    Num_sol = np.insert(Num_sol, 0, alpha)
    Num_sol = np.insert(Num_sol, len(Num_sol), beta)    


    solution = np.sin(grid_pts)

#     plt.plot(grid_pts, Num_sol, 'r')
#     plt.plot(grid_pts, solution, 'b')
#     plt.show()
    
    return(Num_sol, solution)
```

```python
a = 0
b = 2*np.pi
alpha = np.sin(a)
beta = np.sin(b)

num_sol_1, sol_1 = second_ord(a, b, alpha, beta, f, 10, tridiagonal)
num_sol_2, sol_2 = second_ord(a, b, alpha, beta, f, 100, tridiagonal)
num_sol_3, sol_3 = second_ord(a, b, alpha, beta, f, 200, tridiagonal)
num_sol_4, sol_4 = second_ord(a, b, alpha, beta, f, 500, tridiagonal)


error_1 = max(abs(num_sol_1 - sol_1))
error_2 = max(abs(num_sol_2 - sol_2))
error_3 = max(abs(num_sol_3 - sol_3))
error_4 = max(abs(num_sol_4 - sol_4))


slope_1 = (np.log(error_2) - np.log(error_1))/(np.log(1/101) - np.log(1/11))
slope_2 = (np.log(error_3) - np.log(error_2))/(np.log(1/201) - np.log(1/101))
slope_3 = (np.log(error_4) - np.log(error_3))/(np.log(1/501) - np.log(1/201))

print('--------------------------------------------------------')
print('Maximum Error')
print('--------------------------------------------------------')
print('m       h          error')
print('10      {:.4f}'.format(1/11), '   ', error_1)
print('100     {:.4f}'.format(1/101), '   ', error_2)
print('200     {:.4f}'.format(1/201), '   ', error_3)
print('500     {:.4f}'.format(1/501), '   ', error_4)

print('--------------------------------------------------------')
print('Error Rate')
print('--------------------------------------------------------')
print('m       h          Error Rate')
print('10      {:.4f}'.format(1/11), '   ')
print('100     {:.4f}'.format(1/101), '   ', slope_1)
print('200     {:.4f}'.format(1/201), '   ', slope_2)
print('500     {:.4f}'.format(1/501), '   ', slope_3)


errors = np.array([error_1, error_2, error_3, error_4])
increment = np.array([1/11, 1/101, 1/201, 1/501])
m, c = np.polyfit(np.log(increment), np.log(errors), 1)

print()
print()

print('The error rate using Linear Regression is', m)
```

## Green's Function 

If we consider the modified system has the form $ AU = F $, where now

\begin{align*}
A = \frac{1}{h^2} \begin{bmatrix}
    h^2          & 0            & \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{\cdots} & \hphantom{0} \\
    1            & -2           & 1            & \hphantom{0} & \hphantom{0} & \hphantom{\cdots} & \hphantom{0} \\
    \hphantom{0} & 1            & -2           & 1            & \hphantom{0} & \hphantom{\cdots} & \hphantom{0} \\
    \hphantom{0} & \hphantom{0} & \ddots       & \ddots       & \ddots       & \hphantom{0}      & \hphantom{0} \\
    \hphantom{0} & \hphantom{0} & \hphantom{0} & 1            & -2           & 1                 & \hphantom{0} \\
    \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{0} & 1            & -2                & 1            \\
    \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{0} & 0                 & h^2
\end{bmatrix}, \hphantom{\cdots}
    U = \begin{bmatrix}
        U_0 \\
        U_1 \\
        U_2 \\
        \vdots \\
        U_{m-1} \\
        U_m \\
        U_{m+1} \\
    \end{bmatrix}, \hphantom{\cdots}
    F = \begin{bmatrix}
        \alpha \\
        f(x_1) \\
        f(x_2) \\
        \vdots \\
        f(x_{m-1}) \\
        f(x_m) \\
        \beta \\
    \end{bmatrix}
\end{align*}

Let $B$ denote the $(m+2) \times (m+2)$ inverse of A. We will index the elements of $B$ by $B_{00} \text{ through } B_{m+1,m+1}$, in the obvious manner. The of $B$ are
\begin{align*}
    &B_{i0}   = G_0(x_i) = 1 - x_i, \text{ for } i = 0, 1, \ldots, m + 1 \\ 
    &B_{im+1} = G_1(x_i) = x_i, \text{ for } i = 0, 1, \ldots, m + 1 \\
    &\text{For the interior columns } (1 \leq j \leq m) \\
    &B_{ij} = hG(x_i:x_j) = \begin{cases}
        h(x_j - 1)x_i, \hphantom{for} i = 1, 2, \ldots, j, \\
        h(x_i - 1)x_j, \hphantom{for} i = j, j+1, \ldots, m.
    \end{cases}  
\end{align*}

Therefore the solution $U = BF$ is

\begin{align*}
    U &= \alpha B_0 + \beta B_{m+1} + \sum_{j = 1}^{m} f_jB_j\\
    U_i &= \alpha (1-x_i) + \beta x_i + h\sum_{j = 1}^m f_jG(x_i:x_j)
\end{align*}




```python
# F_2 = np.array([alpha])
# for i in range(1, m+1):
#     F_2 = np.insert(F_2, len(F_2), f(a+i*h))
    
# F_2 =  np.insert(F_2, len(F_2), beta)


def green_fun_method(a, b, alpha, beta, f, m):
    h = (b - a)/(m+1)
    B = np.zeros((m+2, m+2))
    grid_pts = np.linspace(a, b, m + 2)
    F = f(grid_pts)
    F[0] = alpha 
    F[-1] = beta

    for i in range(m+2):
        for j in range(m+2):
            if j == 0:
                B[i, j] = 1 - grid_pts[i]
            elif j == m+1:
                B[i, j] = grid_pts[i]
            else:
                if i <= j:
                    B[i, j] = h*(grid_pts[j] - 1)*grid_pts[i]
                else:
                    B[i, j] = h*grid_pts[j]*(grid_pts[i] - 1)
                    
    Num_sol = np.dot(B, F)
    return(Num_sol)
```

## Neumann Boundary conditions

Here we consider with one Neumann condition, say
\begin{align*}
    u'(0) = \sigma, \hphantom{sp} u(1) = \beta
\end{align*}

There are three approaches,

### First approach
We will consider one-sided expression for $u'(0)$ such as
\begin{align*}
    \frac{U_1 - U_0}{h} = \sigma
\end{align*}

If we use this equation in place of the first line of the system in the previous block, then we obtain the following system of equations for the unknowns $U_0, U_1, \ldots, U_m, U_{m+1}:$

\begin{align*}
\frac{1}{h^2} \begin{bmatrix}
    -h            & h            & \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{\cdots} & \hphantom{0} \\
    1            & -2           & 1            & \hphantom{0} & \hphantom{0} & \hphantom{\cdots} & \hphantom{0} \\
    \hphantom{0} & 1            & -2           & 1            & \hphantom{0} & \hphantom{\cdots} & \hphantom{0} \\
    \hphantom{0} & \hphantom{0} & \ddots       & \ddots       & \ddots       & \hphantom{0}      & \hphantom{0} \\
    \hphantom{0} & \hphantom{0} & \hphantom{0} & 1            & -2           & 1                 & \hphantom{0} \\
    \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{0} & 1            & -2                & 1            \\
    \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{0} & 0                 & h^2
\end{bmatrix}
    \begin{bmatrix}
        U_0 \\
        U_1 \\
        U_2 \\
        \vdots \\
        U_{m-1} \\
        U_m \\
        U_{m+1} \\
    \end{bmatrix} = 
    \begin{bmatrix}
        \alpha \\
        f(x_1) \\
        f(x_2) \\
        \vdots \\
        f(x_{m-1}) \\
        f(x_m) \\
        \beta \\
    \end{bmatrix}
\end{align*}

But checking the errors shows that this is only first order accurate

### Second approach
To obtain a second order accuarate method, we can use a centered approximation to $u'(0) = \sigma$ instead of the one-sided approximation. Introducing $U_{-1}$ and,instead of the single equation, we use the following two equations:
\begin{align*}
    \frac{1}{h^2}(U_{-1} - 2U_0 + U_1 &= f(x_0) \\
    \frac{1}{h^2}(U_1 - U_{-1}) &= \sigma
\end{align*}

We can eliminate the unknown $U_{-1}$ and get the following equation

\begin{align*}
    \frac{1}{h}(-U_0 + U_1) = \sigma + \frac{h}{2}f(x_0)
\end{align*}

### Third approach
Instead of using second order center approximation we will use second order forward approximation
\begin{align*}
    \frac{1}{h^2}\left(-\frac{3}{2}U_0 + 2U_1 - \frac{1}{2}U_2 \right) = \sigma
\end{align*}

This results in the linear system

\begin{align*}
\frac{1}{h^2} \begin{bmatrix}
    -\frac{3}{2}h & 2h          & -\frac{1}{2}h & \hphantom{0} & \hphantom{0} & \hphantom{\cdots} & \hphantom{0} \\
    1            & -2           & 1            & \hphantom{0} & \hphantom{0} & \hphantom{\cdots} & \hphantom{0} \\
    \hphantom{0} & 1            & -2           & 1            & \hphantom{0} & \hphantom{\cdots} & \hphantom{0} \\
    \hphantom{0} & \hphantom{0} & \ddots       & \ddots       & \ddots       & \hphantom{0}      & \hphantom{0} \\
    \hphantom{0} & \hphantom{0} & \hphantom{0} & 1            & -2           & 1                 & \hphantom{0} \\
    \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{0} & 1            & -2                & 1            \\
    \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{0} & \hphantom{0} & 0                 & h^2
\end{bmatrix}\begin{bmatrix}
        U_0 \\
        U_1 \\
        U_2 \\
        \vdots \\
        U_{m-1} \\
        U_m \\
        U_{m+1} \\
    \end{bmatrix} = 
    \begin{bmatrix}
        \alpha \\
        f(x_1) \\
        f(x_2) \\
        \vdots \\
        f(x_{m-1}) \\
        f(x_m) \\
        \beta \\
    \end{bmatrix}
\end{align*}

```python
# Coefficient matrix of the first approach
# Find the convergence rate

# Coeff_app_1 = []

# for i in range(m+2):
#     Coeff_app_1.append(0)

def Neumann_fa(a, b, sigma, beta, f, m, tridiagonal):
    h = (b - a)/(m+1)
    
    grid_pts = np.linspace(a, b, m + 2)

    Coeff_app_1 = tridiagonal(np.full(m + 2, -2.0), np.full(m + 1, 1.0), np.full(m + 1, 1.0))
    Coeff_app_1[0, 0] = -h
    Coeff_app_1[0, 1] = h

    Coeff_app_1[-1, -1] = h*h
    Coeff_app_1[-1, -2] = 0
    
    F = f(grid_pts)
    F[-1] = beta
    F[0] = sigma

    return(np.linalg.solve(Coeff_app_1, h*h*F))

def Neumann_sa(a, b, sigma, beta, f, m, tridiagonal):
    h = (b - a)/(m+1)
    
    grid_pts = np.linspace(a, b, m + 2)

    Coeff_app_1 = tridiagonal(np.full(m + 2, -2.0), np.full(m + 1, 1.0), np.full(m + 1, 1.0))
    Coeff_app_1[0, 0] = -h
    Coeff_app_1[0, 1] = h

    Coeff_app_1[-1, -1] = h*h
    Coeff_app_1[-1, -2] = 0

    # Coefficient matrix of the second approach is same but the rhs matrix is different
    F = f(grid_pts)
    F[0] = sigma + h/2 * f(a)
    F[-1] = beta
    
    return(np.linalg.solve(Coeff_app_1, h*h*F))


def Neumann_ta(a, b, sigma, beta, f, m, tridiagonal):
    # Coefficient matrix of the third approach
    h = (b - a)/(m + 1)

    Coeff_app_1 = tridiagonal(np.full(m + 2, -2.0), np.full(m + 1, 1.0), np.full(m + 1, 1.0))
    Coeff_app_1[0, 0] = -1.5*h
    Coeff_app_1[0, 1] = 2*h
    Coeff_app_1[0, 2] = -0.5*h

    Coeff_app_1[-1, -1] = h*h
    Coeff_app_1[-1, -2] = 0

    # print(Coeff_app_3)
    # print(np.array(Coeff_app_1))
    F = f(np.linspace(a, b, m + 2))
    F[0] = sigma
    F[-1] = beta
    

    return(np.linalg.solve(Coeff_app_1, h*h*F))

```

## A general linear second order equation
We now consider the more general linear equation
\begin{align*}
    a(x)u''(x) + b(x)u'(x) + c(x)u(x) = f(x),
\end{align*}
together with Dirichlet boundary conditions
\begin{align*}
    u(a) = \alpha, \hphantom{spa} u(b) = \beta.
\end{align*}

This equation can be discretized to second order by
\begin{align*}
    a_i\left(\frac{U_{i-1} - 2U_i+U_{i+1}}{h^2} \right) + b_i\left(\frac{U_{i+1} - U_{i-1}}{2h} \right) + c_iU_i= f_i,
\end{align*}
where a_i = a(x_i). This gives the linear system $AU=F$, where
\begin{align*}
    A = \frac{1}{h^2}\begin{bmatrix}
        h^2c_1 - 2a_1 & a_1 + hb_1/2         & \hphantom{0}          & \hphantom{0}       \\
        a_2 - hb_2/2  & h^2c_2 - 2a_2        & a_2 + hb_2/2          & \hphantom{0}       \\
        \ddots        & \ddots               & \ddots                & \hphantom{0}       \\
        \hphantom{0}  & a_{m-1} - hb_{m-1}/2 & h^2c_{m-1} - 2a_{m-1} & a_{m-1}+hb_{m-1}/2 \\
        \hphantom{0}  & \hphantom{0}         & a_m - hb_m/2          & h^2c_m-2a_m
    \end{bmatrix}, \hphantom{a} U = \begin{bmatrix}
        U_1 \\
        U_2 \\
        \vdots \\
        U_{m-1} \\
        U_m
    \end{bmatrix} \text{ and } F = \begin{bmatrix}
        f_1 - (a_1/h^2 - b_1/2h)\alpha \\
        f_2 \\
        \vdots \\
        f_{m-1} \\
        f_m - (a_m/h^2 + b_m/2h)\beta
    \end{bmatrix}
\end{align*}

```python
# Coefficient matrix
# Convergence rate

def linear_solution(a, b, alpha, beta, f, a_func, b_func, c_func, m, tridiagonal):
    h = (b - a)/(m+1)

    grid_pts = np.linspace(a, b, m + 2)
    a = a_func(grid_pts[1:m+1])
    b = b_func(grid_pts[1:m+1])
    c = c_func(grid_pts[1:m+1])
        
    Coeff_gen = tridiagonal(h*h*c - 2*a, a[:m - 1] - h*b[: m - 1]/2, a[1:] - h*b[1:]/2 )

    F_gen = f(grid_pts[1: m+1])
    
    F_gen[0] = F_gen[0] - (a[0]/h**2 - b[0]/(2*h))*alpha
    F_gen[-1] = F_gen[m-1] - (a[m-1]/h**2 - b[m-1]/(2*h))*beta
    

    Num_sol_gen = np.linalg.solve(Coeff_gen, h*h*F_gen)

    Num_sol_gen = np.insert(Num_sol_gen, 0, alpha)
    Num_sol_gen = np.insert(Num_sol_gen, len(Num_sol_gen), beta)
    
    return(Num_sol_gen)
#     print(Num_sol_gen)
#     print(Num_sol_1)
```

This is a second order discretization but may not be good for certain problems, for example if we consider the below equation
\begin{align*}
    (\kappa (x) u'(x))' = f(x)
\end{align*}
together with Dirichlet boundary condition
We will not discretize 
\begin{align*}
    \kappa (x) u''(x) + \kappa '(x)u'(x) = f(x)
\end{align*}

Rather we will first approximate $\kappa (x)u'(x)$ at points halfway between the grid points, using a centered approximation

\begin{align*}
    \kappa (x_{i+1/2})u'(x_{i + 1/2}) = \kappa_{i + 1/2}\left(\frac{U_{i+1} - U_i}{h}\right)
\end{align*}
and the analogous approximation at $x_{i - 1/2}$. Differencing these then gives a centered approximation to $(\kappa u')'$ at the grid point $x_i$:
\begin{align*}
    (\kappa u')' & \approx \frac{1}{h} \left[ \kappa_{i + 1/2} \left(\frac{U_{i+1} - U_i}{h}\right) - 
    \kappa_{i - 1/2} \left( \frac{U_i - U_{i - 1}}{h} \right) \right] \\
    & = \frac{1}{h^2} \left[ \kappa_{i - 1/2}U_{i - 1} - (\kappa_{i - 1/2} + \kappa_{i + 1/2})U_i + 
    \kappa_{i + 1/2}U_{i+1} \right].
\end{align*}
Therefore the coefficient matrix is
\begin{align*}
    A = \frac{1}{h^2}\begin{bmatrix}
        -(\kappa_{1/2} + \kappa_{3/2}) & \kappa_{3/2}                   & \hphantom{0} & \hphantom{0} \\
        \kappa_{3/2}                   & -(\kappa_{3/2} + \kappa_{5/2}) & \kappa_{5/2} & \hphantom{0} \\
        \ddots                         &  \ddots                        & \ddots       & \hphantom{0} \\
        \hphantom{0}                   & \kappa_{m-3/2} & -(\kappa_{m-3/2} + \kappa_{m-1/2}) & \kappa_{m -1/2} \\
        \hphantom{0}                   & \phantom{0}    & \kappa_{m-1/2} & -(\kappa_{m-1/2} + \kappa_{m+1/2})
    \end{bmatrix}
\end{align*}

```python
def gen_der(x_pts, x_bar, k):
    n = len(x_pts)
    Vand_matrix = np.ones((n, n))
    x_new = x_pts - x_bar
    
    for i in range(1, n):
        Vand_matrix[i] = (x_new)**(i)/math.factorial(i)
        
    
    rhs = np.zeros(n)
    rhs[k] = 1
    
    
    return(np.linalg.solve(Vand_matrix, rhs))


# Below is the coefficient matrix if we have non uniform grid and neumann boundary condition
Coeff_mat = np.zeros((m + 2, m + 2))

non_uni_pts = np.copy(grid_pts)

local_arr = gen_der(non_uni_pts[0:3], non_uni_pts[0], 2)
Coeff_mat[0][0] = local_arr[0]
Coeff_mat[0][1] = local_arr[1]
Coeff_mat[0][2] = local_arr[2]

for i in range(1, m + 1):
    local_arr = gen_der(non_uni_pts[i - 1:i+2], non_uni_pts[i], 2)
    Coeff_mat[i][i - 1] = local_arr[0]
    Coeff_mat[i][i] = local_arr[1]
    Coeff_mat[i][i + 1] = local_arr[2]
    
local_arr = gen_der(non_uni_pts[m : m + 2], non_uni_pts[m + 1], 0)

Coeff_mat[m + 1][m] = local_arr[0]
Coeff_mat[m + 1][m + 1] = local_arr[1]
```

```python
def linear_regression(x, y):
    A = np.zeros((2, 2))
    temp_1 = np.sum(x)
    temp_2 = np.sum(y)
    temp_3 = np.sum(x*x)
    temp_4 = np.dot(x, y)
    n = len(x)
    
    A[0, 0] = n
    A[0, 1] = -temp_1
    A[1, 0] = -temp_1
    A[1, 1] = temp_3
    A = A/(n*temp_3 - temp_1**2)
    
    return(np.dot(A, [temp_4, temp_2]))
```
